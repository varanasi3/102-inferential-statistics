```python
import numpy as np
import matplotlib.pyplot as plt 
import math
import scipy.integrate as integrate
```

```python
def binomial(n,p):
    x = np.arange(n)
    def prob(x):
        return(math.comb(n,x)*(p**x)*(1-p)**(n-x))
    y = []
    for i in range(len(x)) :
        y.append(prob(i))
    print("mean = ", n*p)    
    print("variance = ", n*p*(1-p)) 

    plt.plot(x,y)
    plt.title(" Binomial distribution")
    plt.xlabel("X")
    plt.ylabel("F(x)")
  
    
    plt.show()
```

```python
binomial(20,0.5)
```

# Normal distribution



```python



def normal(mu, sigma):
    x = np.linspace(mu-4*sigma ,mu+4*sigma, 100)
    def prob(x):
        return(1/math.sqrt(2*math.pi*sigma)*math.exp((-1/2)*((x-mu)/sigma)**2))

    
    y = []
    for i in range(len(x)) :
        y.append(prob(i))
        
    print("Mean = ",mu)
    print("Varience = ",sigma)
    plt.bar(x,y)
    plt.title("Normal Distribution")
    plt.xlabel("X")
    plt.ylabel("F(x)")
  
    
    plt.show()
    
    
    
```

```python
normal(50,20)
```

# Poisson($\lambda$)



```python
def poisson(lambdaa):
    x = np.arange(1000)
    def prob(x):
        return((np.exp(-lambdaa))* ((lambdaa**x)/(math.factorial(x))))
    y = []
    for i in range(len(x)) :
        y.append(prob(i))
    print("Mean = ",lambdaa)
    print("Varience = ",lambdaa)

    plt.bar(x,y)
    plt.title(" Poisson distribution")
    plt.xlabel("X")
    plt.ylabel("F(x)")
    plt.xlim(0,20)
  
    
    plt.show()
    
poisson(10)
```

```python
poisson(5)
```

# Gamma distribution


```python
def gamm(alpha):
    return math.factorial(alpha-1)
def f(alpha,lmbda,x):
    return ((math.pow(lmbda,alpha)*math.pow(x,(alpha-1))*math.exp(-lmbda*x))/gamm(alpha))
def gamma(alpha,lmbda):
    X=np.linspace(0,5,1000)
    y = []
    for x in X:
        y.append(f(alpha,lmbda,x))
        plt.grid(True)
      
  
    plt.plot(X,y)
    plt.bar(X,y,color="orange")
    plt.title("Gamma Distribution",fontsize=20)
    plt.xlabel('x')
    plt.ylabel('f(x)')
```
